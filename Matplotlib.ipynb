Matplotlib: useful for making plots

```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
# importing numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
# sine wave
plt.plot(x, y)
plt.show()
```

```python
# cosine wave
plt.plot(x, z)
plt.show()
```

```python
# adding title, x-axis & y-axis
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Wave')
plt.show()
```

```python
# parabola
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'r+')
plt.show()
```

```python
plt.plot(x, y, 'g.')
plt.show()
```

```python
plt.plot(x, y, 'rx')
plt.show()
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

Bar plot

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
ax.bar(languages, popularity)
plt.xlabel('Languages')
plt.ylabel('Popularity')
plt.show()
```

Pie chart

```python
fig1 = plt.figure()
ax1 = fig1.add_axes([0, 0, 1, 1])
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23, 17, 35, 29, 12]
ax1.pie(students, labels = langs)
plt.show()
```

Scatter Plot

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax2 = fig2.add_axes([0, 0, 1, 1])
ax2.scatter(x, y, color='g')
ax2.scatter(x, z, color='b')
plt.show()
```

3D Scatter plot

```python
fig3 = plt.figure()
ax3 = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax3.scatter(x, y, z, c=z, cmap='Greens')
plt.show()
```
